## Plots|

```julia
using Plots
```

```julia
f(x) = 3 * x^2 + 6 * x - 9

for x = -5:5    
  println("(",x, ", ", f(x), ")")
end
```

```julia
gr()           # Activate the GR backend for use with Plots
plot(f, -4, 3) # plot f over [-4,4]

plot!(zero, -4, 3)
```

```julia
f(-3) # is a root
```

```julia
f(1) # is another root
```

```julia
f(-1) # is min.
```
